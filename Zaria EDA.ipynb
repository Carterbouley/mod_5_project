```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('credit_card.xls', header=1, index_col='ID')
```

```python
df.head()
```

```python
df['Gender'] = df['Gender'].astype('category')
df['Education'] = df['Education'].astype('category')
df['MaritalStatus'] = df['MaritalStatus'].astype('category')
```
